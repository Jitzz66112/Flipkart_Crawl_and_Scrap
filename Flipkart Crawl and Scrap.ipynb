# Importing Libraries

```python
import os
import pandas as pd
import numpy as np
import datetime
import time
import dateutil
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup as bs
```

# Crawl and browe parameters

```python
#setting up web browser parameter
options = Options()
options.add_argument("--start-maximized")
#options.add_argument("--headless")
driver = webdriver.Chrome(options = options)

#crawler
driver.get('https://www.flipkart.com/')
time.sleep(5)#page loading

#ignore login
cross = driver.find_element_by_xpath('/html/body/div[2]/div/div/button')
cross.click()
time.sleep(2)

#clicking on search box
detail = driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
detail.click()
time.sleep(2)

#filling product details to search
detail.send_keys('samsung mobiles')
search=driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search.click()
time.sleep(5)

#fetching data
data_dump = driver.page_source
```

```python
data_dump
```

# Scrapping of Data

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(data_dump,'html.parser')#parsing of data
```

# Fetching product name and its price (removing noise)

```python
product_name = soup.findAll('div',{'class':'_3wU53n'})
```

```python
product_name
```

```python
product_name = [x.text for x in product_name]
```

```python
product_name
```

```python
product_price = soup.findAll('div',{'class':'_1vC4OE _2rQ-NK'})
```

```python
product_price
```

```python
product_price = [x.text for x in product_price]
```

```python
product_price
```

```python
data_list = [product_name,product_price]
```

```python
data_list
```

# Creating dataframe and storing the scrapped data into it.

```python
scrapped_data = pd.DataFrame(data=data_list)
```

```python
scrapped_data
```

```python
scrapped_data = scrapped_data.transpose()
```

```python
scrapped_data
```

```python
scrapped_data.columns=['Product Name','Price']
```
